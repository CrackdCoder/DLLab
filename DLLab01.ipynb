```python
import tensorflow as tf
print(tf.__version__)
```

```python
a=tf.constant(10)
print(a)
v=tf.constant([1,2,3,4])
print(v)

m=tf.constant([[1,2],[3,4]])
print(m)

t3=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print(t3)
```

```python
a = tf.constant([10,20,30])
b = tf.constant([40,50,60])

print(a+b)
```

```python
a=tf.constant(24)
b=tf.constant(26)
print(tf.add(a,b))
print(tf.subtract(a,b))
print(tf.multiply(a,b))
print(tf.divide(a,b))
print(tf.pow(a,b))
```

```python
a=tf.constant([10,20,30])
b=tf.constant([34,2,7])
print(tf.add(a,b))
print(tf.subtract(a,b))
print(tf.multiply(a,b))
print(tf.divide(a,b))
```

```python
a= tf.constant([10,20,30])
print(a.dtype,a.shape,tf.rank(a))
```
